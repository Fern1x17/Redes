```python
import requests
import socket
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://httpbin.org/html')

print("Status Code:", response.status_code)
print("Headers:", response.headers)
print("Response Text:", response.text)
```

```python
response = requests.get('https://httpbin.org/html2')

print("Status Code:", response.status_code)
print("Headers:", response.headers)
print("Response Text:", response.text)
print("En este caso, el código de estado es 404, lo que indica que la página no fue encontrada. Esto se debe a que la URL solicitada no existe en el servidor.")
```

```python
response = requests.get('https://httpbin.org/forms/post')

print("Status Code:", response.status_code)
print("Headers:", response.headers)
print("Response Text:", response.text)
```

```python
response = requests.post('https://httpbin.org/post')

print("Status Code:", response.status_code)
print("Headers:", response.headers)
print("Response Text:", response.text)
```

```python
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect(('httpbin.org', 80))
```

```python
request = "GET /html HTTP/1.1\r\nHost: httpbin.org\r\n\r\n"
s.send(request.encode())
response = s.recv(4096)
print("Raw Response:", response.decode())
```

```python
s.close()
```

```python
responses = [requests.get('http://httpbin.org/html'), requests.get('https://httpbin.org/html')]
if responses[0].status_code == 200:
    print("HTTP Response is successful.")
else:    print("HTTP Response is not successful.")
if responses[1].status_code == 200:
    print("HTTPS Response is successful.")
else:    print("HTTPS Response is not successful.")
```

```python
response = requests.get('https://otero-lado-fernando.mooo.com')
soup = BeautifulSoup(response.text, 'html.parser')

parrafos = soup.find_all('p')
print("Numero de párrafos:", len(parrafos))

for tag in soup.find_all(["h1", "h2", "p", "a"]):
    print(tag.text)
```
