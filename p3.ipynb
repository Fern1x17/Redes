```python
import smtplib
import os
from email.mime.text import MIMEText
from dotenv import load_dotenv
import poplib
from email.parser import BytesParser
import ssl
```

```python
load_dotenv("credentials.env")  # Carga las variables de entorno desde el archivo .env
```

```python
smtp_host = os.getenv('SMTP_HOST') # Dirección del servidor SMTP (ejemplo: 'smtp.gmail.com') a traves de credentials.evn
smtp_port = os.getenv('SMTP_PORT') # Puerto 587 para TLS (STARTTLS)
smtp_user = os.getenv('SMTP_USER') # Tu dirección de correo electrónico
smtp_pass = os.getenv('SMTP_PASS') # Tu contraseña de correo electrónico o contraseña de aplicación

# Texto del mensaje
mensaje = MIMEText("Que dame la mandanga y déjame de tema. Dame chocolate que me ponga bien. Dame de la negra que hace buen olor que con la maría valla colocón")
mensaje['Subject'] = "Viva el fary" # Asunto del correo
mensaje['From'] = smtp_user # Emisor
mensaje['To'] = smtp_user # Receptor

# comprobar que hst, port, user y pass no sean None
print("SMTP Host:", smtp_host)
print("SMTP Port:", smtp_port)
print("SMTP User:", smtp_user)
print("SMTP Pass:", smtp_pass)
```

```python
# Apartado 1 con tls

try:
    server = smtplib.SMTP(smtp_host, smtp_port) # iniciar server
    # smtp_host es el servidor SMTP (ejemplo: 'smtp.gmail.com')
    # smtp_port es el puerto (587 para TLS)
    server.set_debuglevel(1)  # Habilita el modo de depuración para ver la comunicación SMTP
    
    #server.ehlo()  # Identifica al cliente con el servidor SMTP
    server.starttls()  # Inicia la conexión TLS
    #server.ehlo()  # Re-identifica al cliente después de iniciar TLS
    
    server.login(smtp_user, smtp_pass)  # Inicia sesión con tus credenciales
    server.send_message(mensaje)  # Envía el mensaje
    server.quit()  # Cierra la conexión
    
    print("Correo enviado con éxito")

except Exception as e:
    print("Error:", e)
```

```python
# Apartado 1 sin tls
smtp_port = 25 # Puerto 25 para SMTP sin TLS

try:
    server = smtplib.SMTP(smtp_host, smtp_port) # iniciar server
    # smtp_host es el servidor SMTP (ejemplo: 'smtp.gmail.com')
    # smtp_port es el puerto (587 para TLS)
    server.set_debuglevel(1)  # Habilita el modo de depuración para ver la comunicación SMTP
    
    server.login(smtp_user, smtp_pass)  # Inicia sesión con tus credenciales
    server.send_message(mensaje)  # Envía el mensaje
    server.quit()  # Cierra la conexión
    
    print("Correo enviado con éxito")

except Exception as e:
    print("Error:", e)
```

```python
pop3_host = os.getenv('POP3_HOST') # Dirección del servidor POP3 (ejemplo: 'pop.gmail.com') a traves de credentials.evn
pop3_port = os.getenv('POP3_PORT') # Puerto 995 para POP3S
pop3_user = os.getenv('POP3_USER') # Tu dirección de correo electrónico
pop3_pass = os.getenv('POP3_PASS') # Tu contraseña de correo electrónico o contraseña

# comprobar que hst, port, user y pass no sean None
print("POP3 Host:", pop3_host)
print("POP3 Port:", pop3_port)
print("POP3 User:", pop3_user)
print("POP3 Pass:", pop3_pass)
```

```python
# Apartado 2 POP3 sin tls
# leer correos con POP3
try:
    # Conectar al servidor POP3
    pop3_server = poplib.POP3(pop3_host, pop3_port)
    # Usamos POP3 en lugar de POP3_SSL porque mailtrap no permite conexiones SSL, pero si tu servidor POP3 requiere SSL, deberías usar poplib.POP3_SSL en su lugar.
    pop3_server.set_debuglevel(1)  # Habilita el modo de depuración para ver la comunicación POP3
    
    pop3_server.user(pop3_user)
    pop3_server.pass_(pop3_pass)    
    # Obtener el número de mensajes
    num_messages = len(pop3_server.list()[1])
    print(f"Número de mensajes: {num_messages}")
    
    # Leer el primer mensaje (si existe)
    if num_messages > 0:
        response, lines, octets = pop3_server.retr(1)  # Obtener el primer mensaje
        msg_content = b"\n".join(lines)  # Unir las líneas del mensaje
        msg = BytesParser().parsebytes(msg_content)  # Parsear el mensaje
        
        print("Asunto:", msg['Subject'])
        print("De:", msg['From'])
        print("Para:", msg['To'])
        payload = msg.get_payload(decode=True)
        print("Contenido:", payload.decode())    
    pop3_server.quit()  # Cerrar la conexión
    
except Exception as e:
    print("Error:", e)
```

```python

# Apartado 2 POP3 con tls y libreria ssl
# leer correos con POP3
pop3_port = 9950 # Puerto 9950 para POP3S con tls
try:
    # Conectar al servidor POP3
    pop3_server = poplib.POP3(pop3_host, pop3_port)
    # Usamos POP3_SSL en lugar de POP3 para conexiones con tls
    pop3_server.set_debuglevel(1)  # Habilita el modo de depuración para ver la comunicación POP3
    
    context = ssl.create_default_context()  # Crear un contexto SSL predeterminado
    
    # Establecer la conexión TLS
    pop3_server.stls(context)  # Iniciar TLS con el contexto SSL
    
    pop3_server.user(pop3_user)
    pop3_server.pass_(pop3_pass)    
    # Obtener el número de mensajes
    num_messages = len(pop3_server.list()[1])
    print(f"Número de mensajes: {num_messages}")
    
    # Leer el primer mensaje (si existe)
    if num_messages > 0:
        response, lines, octets = pop3_server.retr(1)  # Obtener el primer mensaje
        msg_content = b"\n".join(lines)  # Unir las líneas del mensaje
        msg = BytesParser().parsebytes(msg_content)  # Parsear el mensaje
        
        print("Asunto:", msg['Subject'])
        print("De:", msg['From'])
        print("Para:", msg['To'])
        payload = msg.get_payload(decode=True)
        print("Contenido:", payload.decode())
    
    pop3_server.quit()  # Cerrar la conexión
    
except Exception as e:
    print("Error:", e)
```
